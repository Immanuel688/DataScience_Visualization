#BIVARIATE ANALYSIS: DISTRIBUTION PLOT
here 'x' and 'y' axes are numerical columns and given by us.

```python
sns.jointplot(x = 'ssc_p',y = 'salary',data = dataset)
plt.show()
#SCATTER PLOT
```

```python
#HEAT MAP FASHION
sns.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='hex')
plt.show()
```

```python
sns.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='kde')
plt.show()
```

```python
#PAIR PLOT
sns.pairplot(dataset,hue = 'gender',palette = "husl")
plt.show()
# IT shows only the plot BETWEEN THE QUANTITATIVE COLUMNS FRPM THE DATASET
```

# FOR CATEGORICAL PLOT

```python
sns.stripplot( x= 'ssc_b', y= 'ssc_p', data=dataset, jitter=True, hue= 'status')
```

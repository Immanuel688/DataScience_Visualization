```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "svg"
#Browser compatability and ease of presentation
```

```python
raw= pd.read_csv("Zomato Chennai Listing 2020.csv")
raw
```

```python
#Finding null value
raw.isnull().sum()
```

```python
raw.info()
```

```python
# Replacing null value as NA for 'object' data type
raw['Dining Rating Count'].fillna('NA',inplace=True)
raw['Delivery Rating Count'].fillna('NA',inplace=True)
```

```python
# Replacing null value with 0.0 for float data type
raw['Dining Rating'].fillna(0.0,inplace=True)
raw['Delivery Rating'].fillna(0.0,inplace=True)
```

```python
raw.isnull().sum()
```

```python
raw.info()
```

```python
# Need to convert numerical columns to numerical data types
# REPLACE CATEGORICAL VALUES WITH '0'
```

```python
raw['Dining Rating Count'].value_counts()
```

```python
raw['Dining Rating Count'].replace(to_replace= ['Does not offer Dining','Not enough Dining Reviews','NA'],value=0,inplace=True)
```

```python
raw['Dining Rating Count'].value_counts()
```

```python
raw['Delivery Rating Count'].value_counts()
```

```python
raw['Delivery Rating Count'].replace(to_replace= ['Does not offer Delivery','Not enough Delivery Reviews','NA'],value=0,inplace=True)
```

```python
raw['Delivery Rating Count'].value_counts()
```

```python
raw.isnull().sum()
```

```python
raw.info()
```

```python
raw['Delivery Rating Count'] = raw['Delivery Rating Count'].astype(int)
raw['Dining Rating Count'] = raw['Dining Rating Count'].astype(int)
```

```python
# REMOVE DATA WHERETHE VALUES ARE '0'
useless=(raw['Delivery Rating Count']==0) |(raw['Dining Rating Count']==0) |(raw['Dining Rating']==0) |(raw['Delivery Rating']==0)
raw[useless]
```

```python
raw=raw[~useless]
raw
```

```python
raw.isnull().sum()
```

```python
raw= raw.drop(['Zomato URL','Address'],axis=1)
raw
```

```python
raw['Top Dishes'].value_counts()
```

```python
raw['Location'].isnull().sum()
```

```python
raw['Location'].value_counts()
```
